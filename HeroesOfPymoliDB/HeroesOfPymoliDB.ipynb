```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
players_df = purchase_data["SN"]
uniqueplayers = len(purchase_data["SN"].unique())
print("Total Players: " + str(uniqueplayers))
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
total_purchases = players_df.count()
item_count = len(purchase_data["Item ID"].unique())
avg_price = purchase_data["Price"].mean()
total_rev = purchase_data["Price"].sum()
basic_calcs = pd.DataFrame({"Unique items": [item_count],"Average Price": [avg_price],
                            "Total # of Purchases":[total_purchases],"Total Revenue":[total_rev]})

basic_calcs["Average Price"] = basic_calcs["Average Price"].map("${:.2f}".format)
basic_calcs["Total Revenue"] = basic_calcs["Total Revenue"].map("${:,.2f}".format)
#Need to figure out how to format the avg price & get rid of the 0 in the far left column
basic_calcs
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Demographic calculations
male_counts = purchase_data[purchase_data["Gender"] == "Male"]
males = male_counts["SN"].nunique()
male_purchases = male_counts["Price"].sum()

female_counts = purchase_data[purchase_data["Gender"] == "Female"]
females = female_counts["SN"].nunique()
female_purchases = female_counts["Price"].sum()

other_counts = purchase_data[purchase_data["Gender"] == "Other / Non-Disclosed"]
others = other_counts["SN"].nunique()
other_purchases = other_counts["Price"].sum()

male_percent = males/uniqueplayers
females_percent = females/uniqueplayers
other_percent = others/uniqueplayers
males_avg = male_purchases/males
females_avg = female_purchases/females
others_avg = other_purchases/others
grandtotal_counts = males + females + others

print(males)
print(male_percent)
print(females)
print(females_percent)
print(others)
print(other_percent)
print(male_purchases)
print(female_purchases)
print(other_purchases)
print(males_avg)
print(females_avg)
print(others_avg)

max_age = purchase_data["Age"].max()
min_age = purchase_data["Age"].min()
print(max_age)
print(min_age)
print(grandtotal_counts)
```

```python
#Demographic summary table

gender_summary = pd.DataFrame([(males,male_percent*100),(females, females_percent*100),(others, other_percent*100)],
                               index = ["Males","Females","Others/Non-Disclosed"],
                             columns = ["Total Count","Percentage of Players"])

gender_summary["Percentage of Players"] = gender_summary["Percentage of Players"].map("{:.2f}%".format)
gender_summary
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#create groups based on SN

def my_agg(x):
    names = {
        'Purchase Count': x['Price'].count(),
        'Average Purchase Price':  x['Price'].mean(),
        'Total Purchase Value': x['Price'].sum()}

    return pd.Series(names, index=['Purchase Count', 'Average Purchase Price', 'Total Purchase Value'])

purchase_data.groupby('SN').apply(my_agg).nlargest(5,'Total Purchase Value')
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_itemdf = purchase_data.groupby(['Item ID','Item Name'])
purch_count = popular_itemdf["Item ID"].count()
item_price = popular_itemdf["Price"].mean()
tot_pur_val = popular_itemdf["Price"].sum()

popular_items_summdf = pd.DataFrame({"Item Purchase Count": purch_count,"Item Price": item_price,
                                     "Total Purchase Value": tot_pur_val})
popular_items_summdf["Item Price"] = popular_items_summdf["Item Price"].map("${:.2f}".format)
popular_items_summdf["Total Purchase Value"] = popular_items_summdf["Total Purchase Value"].map("${:.2f}".format)
popular_items_summdf.nlargest(5,'Item Purchase Count')
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
popular_items_summdf = pd.DataFrame({"Item Purchase Count": purch_count,"Item Price": item_price,
                                     "Total Purchase Value": tot_pur_val})



popular_items_summdf = popular_items_summdf.sort_values("Total Purchase Value", ascending=False)
popular_items_summdf["Total Purchase Value"] = popular_items_summdf["Total Purchase Value"].map("${:.2f}".format)
popular_items_summdf["Item Price"] = popular_items_summdf["Item Price"].map("${:.2f}".format)

popular_items_summdf.head(5)
```
